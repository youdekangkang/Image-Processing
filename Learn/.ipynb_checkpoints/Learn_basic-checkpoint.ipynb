# 视频读取

```python
vc = cv2.VideoCapture('test1.mp4')

# 检查是否打开正确
if vc.isOpened():
    open,frame = vc.read()
else:
    open = False
```

```python
while open:
    #frame代表每一个画面 ret代表每一帧
    ret,frame = vc.read()
    if frame is None:
        break
    if ret == True:
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('resule',gray)
        #27代表esc按键
        if cv2.waitKey(10)&0xFF == '27':
            break
vc.release()
cv2.destroyAllWindows()
```

# 边界填充

```python
top_size,bottom_size,left_size,right_size = (50,50,50,50)

replicate = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REPLICATE)
refelct = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_WRAP)
constant = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_CONSTANT,value = 0)
```

* BORDER_REPLICATE  复制最边缘的像素
* BORDER_REFLECT   反射法 对感兴趣图像中的像素在两边进行复制
* BORDER_REFLECT_101 最边缘的像素为轴对称
* BORDER_WRAP     外包装法
* BORDER_CONSTANT  常量填充

import matplotlib.pyplot as plt

plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
